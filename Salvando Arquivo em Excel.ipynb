```python
import pandas as pd
```

```python
tabela = pd.read_excel(r'Produtos.xlsx')
display(tabela)
```

```python
#atualizar o multiplicado onde a coluna tipo da tabela seja == Serviço
tabela.loc[tabela['Tipo']== 'Serviço', 'Multiplicador Imposto'] = 1.5
#atualizando base de preço
tabela['Preço Base Reais'] = tabela['Preço Base Original']*tabela['Multiplicador Imposto']
display(tabela)
```

```python
tabela.to_excel('Novos Produtos.xlsx', index=False)
```

```python
#formatando dados com OpenPyxl
from openpyxl import Workbook, load_workbook
```

```python
planilha = load_workbook('Produtos.xlsx')
```

```python
aba_ativa =  planilha.active
for celula in aba_ativa['C']:
    if celula.value == 'Serviço':
        linha = celula.row
        aba_ativa[f'D{linha}'] = 1.5
planilha.save('produtosOpenPy.xlsx')
```
